## 1. Setup

Before running any examples, let's import the required libraries and initialize models.

```python
import os, sys
from pathlib import Path

os.environ["DYNAMIC_SAM2_PATH"] = str((Path().resolve() / "../../DynamicSAM2").resolve())
```
